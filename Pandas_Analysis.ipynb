Task 2: Handle Missing Values
Clean the dataset using appropriate strategies:

Region & Product_category: Fill with mode (most frequent value)
Sales_amount: Fill with median
Quantity: Fill using forward fill (ffill)
Customer_age: Fill with mean (rounded to integer)
Payment_method: Drop rows where payment_method is missing
Verify no missing values remain using df.isna().sum()

```python
#Fill region with mode
df['region'].fillna(df['region'].mode()[0])
```

```python
#Fill product_category with mode
df['product_category'].fillna(df['product_category'].mode()[0])
```

```python
#Fill sales_amount with median
df['sales_amount'].fillna(df['sales_amount'].median())
```

```python
#Fill quantity using forward fill
#df['quantity'].fillna(method='ffill')
df['quantity'].ffill()
```

```python
#Fill customer_age with mean
#df['customer_age'].fillna(round(df['customer_age'].mean()), inplace=True)
df['customer_age'].fillna(round(df['customer_age'].mean()))
```

```python
#Drop rows where payment_method is missing
df.dropna(subset=['payment_method'])
```

```python
#Verify No Missing Values
df.isna().sum()
```

Task 3 — GroupBy Analysis

```python
#Total Sales by Region
df.groupby('region')['sales_amount'].sum()
```

```python
#Average Sales by Product Category
df.groupby('product_category')['sales_amount'].mean()
```

```python
#Group by Region and Product Category
region_product_sales = df.groupby(
    ['region', 'product_category']
)[['sales_amount', 'quantity']].sum().reset_index()

region_product_sales
```

```python
#Top 3 Region-Product Combinations by Sales
region_product_sales.sort_values(
    by='sales_amount', ascending=False
).head(3)
```

Task 4 — Custom Aggregation

```python
#Create Custom Function
def sales_range(x):
    return x.max() - x.min()
```

```python
#Sales Range by Region
df.groupby('region')['sales_amount'].apply(sales_range)
```

```python
#Multiple Aggregations Using .agg()
df.groupby('region').agg({
    'sales_amount': ['sum', 'mean', 'max'],
    'quantity': ['sum', 'min']
})
```
